```python
from Bio.SeqUtils import ProtParam
import pandas as pd

#reading peptides from file and couting them in % utilizing ProtParam

peptide_data = []

with open('../data/raw/positive.txt') as reader:
    for peptide in reader:
        aa_composition = ProtParam.ProteinAnalysis(peptide).get_amino_acids_percent()
        aa_composition['label'] = 1 #1 for positive
        peptide_data.append(aa_composition)
        
with open('../data/raw/negative.txt') as reader:
    for peptide in reader:
        aa_composition = ProtParam.ProteinAnalysis(peptide).get_amino_acids_percent()
        aa_composition['label'] = 0 #0 for negative
        peptide_data.append(aa_composition)
        
#creating data frame
df_peptides = pd.DataFrame(peptide_data)
df_peptides
```

```python
X = df_peptides.drop(['label'], axis=1) #attributes
y = df_peptides['label'] #prediction, classification
```

```python
from sklearn.decomposition import PCA #reduction of dimension
from plotly import express as ex

pca = PCA(n_components =2) #bidimensional graph
X_pca = pca.fit_transform(X)

df_peptides_pca = pd.DataFrame(X_pca, columns = ['pca_1', 'pca_2'])
df_peptides_pca['label'] = y.astype(str) #change color and numbers in graph

ex.scatter(df_peptides_pca, x='pca_1', y='pca_2', color='label')
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(X_pca, y)
```

```python
from umap import UMAP

umap = UMAP(n_components =2) #bidimensional graph
X_umap = umap.fit_transform(X)

df_peptides_umap = pd.DataFrame(X_umap, columns = ['umap_1', 'umap_2'])
df_peptides_umap['label'] = y.astype(str) #change color and numbers in graph

ex.scatter(df_peptides_umap, x='umap_1', y='umap_2', color='label')
```

```python
silhouette_score(X_umap, y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier 
from sklearn.metrics import classification_report

#spliting data
X_train, X_test, y_train, y_test = train_test_split(X,y)

#model construction
model = RandomForestClassifier()
model.fit(X_train, y_train)

#prediction
y_pred = model.predict(X_test)

report = classification_report(y_test, y_pred)

print(report)
```

```python
#creating function to predict anticancer peptide

def predict_anticancer_peptide(peptide_sequence):
    aa_composition = ProtParam.ProteinAnalysis(peptide_sequence).get_amino_acids_percent()
    X = pd.DataFrame([aa_composition])
    return model.predict_proba(X)[0][1]

predict_anticancer_peptide('WWWWGGGG')
```
